## **Task 1**

### **Explain the concept of portals in React and describe how they differ from regular DOM rendering.**

### **Answer:**

Portals in React allow rendering a component outside its parent DOM hierarchy using `ReactDOM.createPortal`.

In regular DOM rendering, components are rendered inside their parent element.
In portals, components are rendered in a different DOM node, typically used for modals, tooltips, and overlays.

## **Task 2**

### **Create a simple React component that uses a portal to render a modal dialog. Ensure that the modal can be opened and closed using state management.**

### **Answer:**

```javascript id="r2k8pm"
import React, { useState } from "react";
import ReactDOM from "react-dom";

function Modal({ onClose }) {
  return ReactDOM.createPortal(
    <div style={{ position: "fixed", top: "30%", left: "30%", background: "white", padding: "20px" }}>
      <h2>Modal</h2>
      <button onClick={onClose}>Close</button>
    </div>,
    document.getElementById("modal-root")
  );
}

function App() {
  const [show, setShow] = useState(false);

  return (
    <div>
      <button onClick={() => setShow(true)}>Open Modal</button>
      {show && <Modal onClose={() => setShow(false)} />}
    </div>
  );
}

export default App;
```

---

## **Task 3**

### **What is React Fiber, and what are its primary goals? Discuss how it improves the reconciliation process compared to previous versions of React.**

### **Answer:**

React Fiber is the new reconciliation engine in React that improves rendering performance.

Primary goals:

* Enable incremental rendering
* Prioritize updates
* Pause and resume work
* Improve responsiveness

Improvement:
Earlier React processed updates synchronously. Fiber breaks work into units, allowing React to handle complex UI updates efficiently without blocking the main thread.

---

## **Task 4**

### **What is React Suspense? Explain its purpose in managing asynchronous operations in React applications. Write a simple example to illustrate suspense in React**

### **Answer:**

React Suspense is used to handle asynchronous operations like lazy loading components by showing a fallback UI while content loads.

```javascript id="r4x9dn"
import React, { Suspense, lazy } from "react";

const LazyComponent = lazy(() => import("./MyComponent"));

function App() {
  return (
    <Suspense fallback={<h2>Loading...</h2>}>
      <LazyComponent />
    </Suspense>
  );
}

export default App;
```

## **Task 5**

### **List and explain at least five best practices for optimizing performance in React applications. Provide examples where applicable.**

### **Answer:**

1. **Use React.memo**
   Prevents unnecessary re-renders of components

2. **Use useCallback**
   Memoizes functions to avoid recreation

3. **Use useMemo**
   Optimizes expensive calculations

4. **Code Splitting (React.lazy)**
   Loads components only when needed

5. **Avoid unnecessary state updates**
   Keep state minimal and efficient

6. **Key prop in lists**
   Helps React identify elements efficiently

Example:

```javascript id="r5m3qa"
const MemoComponent = React.memo(function Component({ value }) {
  return <div>{value}</div>;
});
```

---
